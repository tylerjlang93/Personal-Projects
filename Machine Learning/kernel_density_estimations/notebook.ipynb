_Based on the outline given in the lecture, show mathematically that the MLE for Gaussian mean and variance parameters are given by_ $\hat \mu = \frac{1}{m} \Sigma^m_{i=1} x^i$ .... $\hat \sigma^2 = \frac{1}{m} \Sigma^m_{i=1} (x^i - \hat \mu)^2$ ... _For this derivation, you will also need to show that these estimates for $\mu$ and $\sigma$ are **maximum.**_

With a Gaussian distribution equation given as: 

$p(x|\mu,\sigma) = \frac{1}{(2\pi)^0.5 \sigma}^{e^{\frac{-1}{2\sigma^2}(x-\mu)^2}}$, 

We can, assuming that a given sample is i.i.d, use MLE to find the parameters that optimize the likelihood of the data falling into the given gaussian of the optimized parameters. In other words, the goal is to find the best guassian distribution that is most likely to fit the data, or find the best parameters of the distribution of which the data is most likely to be part of. 

Thus, we can take the likelihood function, and further take the log of it, to make the math easier to equate. 

$l(\mu,\sigma;D) = log \prod \frac{1}{(2\pi)^.5 \sigma}{e^{\frac{-1}{2\sigma^2}(x-\mu)^2}}$,

which can be simplified as 

$ - \frac{m}{2}log(2\pi) -\frac{m}{2}log(\sigma^2) - \Sigma^m_{i=1} \frac{(x^i - \mu)^2}{2\sigma^2} $

Then, to MAXIMIZE this problem, we can simply take the partial derivative with respect to our parameters, $\mu$ and $\sigma$

## For $\mu$:

Taking the derivative, we get: 

$\frac{dl}{d\mu} = 0 - 0 - \Sigma^m_{i=1} \frac{(2)(x^i - \mu)(-1)}{2\sigma^2} = 0$

Now, can simplify by crossing out the 2s in the numerator and the denominator, and cancel out the two negative signs: the one in front of the sum, and the one due to derivation on the right in the numerator:

$ 0 = \Sigma^m_{i=1} \frac{x^i-\mu}{\sigma^2}$

Then, we can take $\frac{1}{\sigma^2}$ out of the summation, and put it on the left, leading to 

$0 = \frac{1}{\sigma^2} \Sigma^m_{i=1}(x^i - \mu)$

Then, multiplying both sides by $\sigma^2$, which we can do because, for this derivative, sigma is considered a constant... and also splitting up the sum into two sums, we get: 

$0 = \Sigma^m_{i=1} x^i - \Sigma^m_{i=1} \mu$

Rearranging by putting the sum of mu on the LHS, we get 

$\Sigma^m_{i=1} x^i = \Sigma^m_{i=1} \mu$.

Now, since mu is our variable, and we are taking m sums of it, we can simply take the mu out of the sum, and multiply it by m, due to basic summation rules! Thus, we get: 

$\Sigma^m_{i=1} x^i = m*\mu$, which we can divide both sides by "m" to get the final equation: 

$\hat \mu = \frac{1}{m} \Sigma^m_{i=1} x^i$

## For $\sigma^2$

Remember our equation: $ - \frac{m}{2}log(2\pi) -\frac{m}{2}log(\sigma^2) - \Sigma^m_{i=1} \frac{(x^i - \mu)^2}{2\sigma^2} $

Now, we can simplify the equation a bit further, comining the two log terms of 2pi and sigma-squared. And then we can also take the $\frac{1}{2\sigma^2}$ out of the sum:

$\frac{dl}{d\sigma^2}[ -(\frac{m}{2}log(2\pi\sigma^2) - \frac{1}{2\sigma^2} \Sigma^m_{i=1} (x^i - \mu)^2)]$


To make this derivation easier, we can first use the product rule, getting us 

$-\frac{m}{2} \frac{dl}{d\sigma^2}[log(s\pi\sigma^2)] + \frac{dl}{d\sigma^2}[-\frac{1}{2\sigma^2} \Sigma^m_{i=1} (x^i - \mu)^2]$

And then, using the chain rule with the log, we can simplify to: 

$-\frac{m}{2\sigma^2} + \frac{dl}{d\sigma^2}[-\frac{1}{2\sigma^2}\Sigma^m_{i=1} (x^i - \mu)^2]$

Finally, to solve for the derivative, we get: 

$0 = \frac{1}{2\sigma^2}(-m + \frac{1}{\sigma^2} \Sigma^m_{i=1} (x^i - \mu)^2)$

Then, rearranging, we get: 

$m\sigma^2 = \Sigma^m_{i=1}(x^i-\mu)^2$,

which simplifies to $\hat \sigma^2 = \frac{1}{m} \Sigma^m_{i=1} (x^i - \hat \mu)^2$

---------------------------

### Proving a MAXIMUM

Now, the big question is, how to know if this is a *maximum or a minimum*.

To know so, we need to use the second-derivative test to get a Hessian matrix, which, if we can prove is negative semi-definite, then we would have a maximum.

Taking our new derivatives, we can then perform the second partial derivative, as such: 

Note, that $\sigma^2$ will be considered a variable, not a squared variable, so taking the derivative will not consider the squared-term, but will take $\sigma^2$ to be a reference to the variance.

#### Taking $\frac{dl}{d\mu^2}$ of $\frac{1}{m} \Sigma^m_{i=1} x^i -  \mu $ 

= $-1$

#### Taking $\frac{dl}{d\mu\sigma^2}$ of $\frac{1}{m} \Sigma^m_{i=1} x^i - \mu $ 
$= 0$

#### Taking $\frac{dl}{d\sigma^4}$  of $\frac{1}{m} \Sigma^m_{i=1} (x^i -  \mu)^2 - \sigma^2$:

$-1$

#### Taking $\frac{dl}{d\sigma^2\mu}$  of $\frac{1}{m} \Sigma^m_{i=1} (x^i - \mu)^2 - \sigma^2$:

$\frac{1}{m} \Sigma^m_{i=1} -2(x^i - \mu)$

As can be seen, all these partial second derivatives would result in a non-positive value! After taking each second-partial-derivative, the only second-partial with any variables left if the derivative in respect to mu, of the derivative in respect to the variance. And in this case, due to the squared term always making the right part of the product a positive value, and then multiplying that value by -2, we will always get a negative number. ## Question 2: [5 points]

_Please compare the pros and cons of KDE as opposed to histograms, and give at least one advantage and disadvantage of each._

### KDE: 
#### Advantages: 

KDEs have many advantages. One advantage is that it is not assumption-heavy, unlike the Gaussian. They are very flexible, do not require learning, and can sometimes better represent data that does not follow a perfect distribution of the many standard distributions we know in Statistics. 

Another advantage of KDE _over histograms_ is that they have lower risk for the same number of samples, or in other words, risk decays faster than with a histogram as the amount of data points increases. In this case, risk is measured by $r(\hat p,p) = \int E_X [(\hat p(x) - p(x))^2] dx $.

#### Disadvantages: 

However, KDEs also come with a price. They are computationally expensive to compute, as they require processing on every single data point, and thus require a lot of memory, unlike a parametric distribution, which just needs a simple list of parameters to quickly form the distribution. Futhermore, while they don't require learning, every time we add or test a new data point, we have to re-run the entire KDE algorithm, with every data point, unlike with a histogram, where we just have to find the appropriate bin and update the count/density. Furthermore, the shape or equation of the KDE is highly dependent on the chosen kernel bandwidth, of which there is no single guaranteed method to pick the "best kernel."


### Histograms:
#### Advantages:

Histograms, like KDEs, are also very flexible, with no strong assumptions necessary. The number of parameters is linked to the number of bins in the model, and not necessarily with the amount of data, unlike KDE, which makes them less memory-intensive and less computationally expensive. 

#### Disadvantages:

Histograms, however, are not smooth, and the shape of a histogram is highly dependent on the amount of bins, or also the bin width, assigned to the histogram. They also do require more parameter storage than a parametric distribution, depending on the number of bins.

Finally, due to the nature of the bins, for every variable, or dimension, in our data, the size of the histogram increases by (quantity of bins per variable)^n! In other words, with just a few variables, and a small amount of bins, the size of the histogram can get extremely big, which is unfeasible for analysis.

## Question 3: [5 points]

_For the EM algorithm for GMM, please show how to use Bayes rule to drive $\tau^i_k$ in closed-form-expression_

For the EM algorithm, we use log-likelihood function to learn the model, and decide, for each data point, which gaussian component it might belong to. However, to perform this log-likelihood, we would have to know the latent factor, z, which is the mixture component, determining which component a point belongs to, with probability $\pi$. Since we do not know Z, we have to maximize **expectations** of the values with respect to the unknown Z, given the name, "Expectation Maximization." 

Specifically, we need to calculate the expectation of the **posterior**, "q", which is the $P(z|x)$. Since we cannot evaluate that directly, we need to use Baye's Rule, which allows us to compute the expectation with values and expressions we can better evaluate. 

As a reminder, Bayes Rule States the following: 

$P(z|x) = \frac{P(x|z)P(z)}{P(x)} = \frac{P(x,z)}{\Sigma_{Z'}P(x,z')}$

In this case, P(x) is the normalization constant, P(x|z) is the likelihood, $=N(x|\mu_z,\Sigma_z)$, and P(z) is the prior, which $=\pi_z$.

Thus, $p(z|x) = $ = $\frac{\pi_z N(x|\mu_z,\Sigma_z)}{\Sigma_{z'} \pi_{z'} N(z|\mu_{z'},\Sigma_{z'})}$.

We find this posterior distribution as $q(z^1,...z^m) = \prod p(z^i = k | x^i, \sigma^t)$ 

Then, computing $p(z^i = k | x^i)$ for each k, we get $\tau^i_k$!



$\tau^i_k$ = $\frac{p(x^i|z^i=k)p(z^i=k)}{\Sigma_{k'=1...K} p(z^i=k',x^i)}$,

which we can simplify to $\tau^i_k = \frac{\pi_k N(x^i | \mu_k,\Sigma_k)}{\Sigma_{k' = 1...K} \pi_{k'} N(x^i | \mu_{k'},\Sigma_{k'})}$

Thus, in summary, we use Bayes Rule because we have a variable, z, of which we do not know its value. Thus, we cannot directly calculate the conditional probability of p(z|x). Thus, using Bayes Rule, we have a work-around, in which we can substitue an equation with three equatable values: p(x|z), p(z), and p(x), and use this function now as the posterior, with relation to theta. Then, finally, we can compute the probability of z_i = k given x, using the substitute equation via Bayes Rule. We call this probability function $\tau$.

## Question c

```python
def KDE_plotter_2D(x,y,xtitle = 'x',ytitle = 'y', title = 'KDE: 2D', figsize = (16,16)):
    # https://towardsdatascience.com/simple-example-of-2d-density-plots-in-python-83b83b934f67
    xmin = x.min()
    xmax = x.max()
    ymin = y.min()
    ymax = y.max()

    delta_x = (max(x) - min(x))/10
    delta_y = (max(y) - min(y))/10
    xmin = min(x) - delta_x
    xmax = max(x) + delta_x
    ymin = min(y) - delta_y
    ymax = max(y) + delta_y
    
    """
    In the above cell, mgrid creates a 100 by 100 matrix, or "mesh" 
    of values around the space of xmin and xmax, to then be used in the part below:
    """
    xx, yy = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]

    positions = np.vstack([xx.ravel(), yy.ravel()])
    values = np.vstack([x, y])
    kernel = st.gaussian_kde(values)
    f = np.reshape(kernel(positions).T, xx.shape)
    """
    Here, the original x and y values are being passed into the gaussian_kde function, 
    which creates an object that can then be used to transform the dense meshgrid we created 
    above, using the distribution created from our original data, which can then be plotted 
    to get the distinct KDE shape we see in the graph below.  
    """
    fig = plt.figure(figsize=figsize)
    ax = fig.gca()
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)
    cfset = ax.contourf(xx, yy, f, cmap='coolwarm')
    ax.imshow(np.rot90(f), cmap='coolwarm', extent=[xmin, xmax, ymin, ymax])
    cset = ax.contour(xx, yy, f, colors='k')
    ax.clabel(cset, inline=1, fontsize=10)
    ax.set_xlabel(xtitle)
    ax.set_ylabel(ytitle)
    plt.title(title)
    plt.scatter(x,y,color='red')
    plt.hlines(y=0,xmin=-1,xmax=1,linestyles='dashed',color='purple',alpha=.5)
    plt.vlines(x=0,ymin=-1,ymax=1,linestyles='dashed',color='purple',alpha=.5)
    return xx,yy,f
```

For this question, we are now asked to use KDE to estimate the 2D density function of amygdala and acc. I will use scipy.stat's great gaussian_kde function, which will give me the values to be plotted and used in the final plotting function, pyplot's contour function, which provides a colored-contour plot of the kernel distribution.

The majority of the code involved in calculating the graph's extremas with the delta_x/y functions, creating an mgrid of values, using scipy.stat's gaussian_kde function on the x and y values, applying this to our positional values to be used in the plot's makeup, and setting up the plot with the coordinate bounds comes from code snippets from

```python
plt.rcParams['figure.figsize'] = [12, 7]
```

```python
xx,yy,f = KDE_plotter_2D(df['amygdala'],df['acc'],'Amygdala','ACC','2D KDE')
```

### Observations: Independent or Dependent?

In the question, we are asked to check if the distributions for the amygdala and acc measurements are independent or not. One way to test independence, is by using the property of joint distributions: That a joint distribution of two random variables is equal to the multiplication of the two marginal distribution *ONLY IF the variables are independent.* Thus, using this property, I can create a function of the distributions of each variable, and then a function of their joint distribution. It is easy, using the function "scipy.stats.gaussian_kde", in which you only have to pass in the data points, in this case the series of amygdala and acc points, and it will create a function that, when passed in an array of new data points, it will output the results using the gaussian kde function created by the original data points. 

Thus, I create one marginal function trained on the amygdala data, one marginal on the acc, and a join gaussian_kde trained on amygdala and acc. Then, I use np.mgrid to create 1D arrays of data, 10,000 points big each, ranging from -.1 to .1 (around the ranges of the original data), one to be passed in to the amygdala kde, and the other into the acc kde, and finally passing both in, as a 2x10000 array, into the joint distribution. 

Then, using the property I mentioned above, I can plot the value of the multiplied marginal distributions - the joint distribution values, and if they are truly equal, the result should be 0 across all results. I then plot the results using plt.imshow, to provide a heatmap of the results, to look for any points in which the two marginals multiplied minus the joint doesn't equal 0. 

```python
marg_amyg = st.gaussian_kde(df['amygdala'])
marg_acc = st.gaussian_kde(df['acc'])
joint = st.gaussian_kde(np.vstack([df['amygdala'], df['acc']]))
```

```python
grid_amyg =np.mgrid[-.1:.1:10000j]
grid_acc  =np.mgrid[-.1:.1:10000j]
```

```python
indep = marg_amyg(grid_amyg)*marg_acc(grid_acc) - joint(np.stack((grid_amyg,grid_acc)))
```

```python
plt.rcParams['figure.figsize'] = [5, 5]
plt.imshow(indep.reshape((100,100)))
plt.colorbar()
plt.show()
```

As can be seen, there is a clear band, or range of values, for which the result does not equal 0, meaning that the two distributions **Are NOT independent!**

```python
amyg_means = []
acc_means = []
orients = []

for i in range(2,df['orientation'].max()+1):
    orients.append(i)
    for reg in ['amygdala','acc']:
        if reg == 'amygdala':
            amyg_means.append(df[df['orientation']==i][reg].mean())
        if reg == 'acc':
            acc_means.append(df[df['orientation']==i][reg].mean())

print("\nThe Conditional Sample Means per Orientation per Brain Region:")
pd.DataFrame(zip(orients,amyg_means,acc_means),columns=['orientation','amygdala','acc']).set_index('orientation').T
```

For this question, we are asked to implement the EM algorithm for fitting a GMM for the MNIST dataset, using the photos for the digits 2 and 6.
```python
plt.rcParams['figure.figsize'] = [10, 7]
def decompose(df,comps,style='svd'):
    
    # Half of this code is attributed directly to the sample code.
    
    assert style in ['svd','sklearn']
    dft = df.T
    m,n = dft.shape
    df_scaled = preprocessing.scale(dft)
    if style == 'sklearn':
        pca = PCA(n_components=comps)
        return pca.fit_transform(df_scaled), m
    if style == 'svd':
        C = np.matmul(df_scaled.T,df_scaled)/m
        V,S,U = np.linalg.svd(C)
        V = V[:,:comps]
        return np.dot(dft,V), m, [V,S,U]
```

## Question a - Written description of E-/M-steps 

Now that I have my data, here I will detail what will be done in part b, in both the E and M steps of the EM algorithm. 


### The E-Step:

For the E-step, after figuring out our posterior distribution using Bayes Rule, and substituting $p(z|x)$ for the likelihood, prior, and normalization constant (   $\frac{P(x|z)P(z)}{P(x)} $   ) (See  **part one** for an explanation of why we do this, and how we come about getting to Bayes Rule)...


We solve for the posterior as $\frac{\pi_z N(x|\mu_z,\Sigma_z)}{\Sigma_{z'} \pi_{z'} N(z|\mu_{z'},\Sigma_{z'})} $, which, written out more fully (considering all values of z1 through zm), we get a product function:

$q(z^1,...z^m) = \prod p(z^i | x^i, \theta^t)$ 


Now, finally, we can move on with the calculation of setting up our equation for maximization. 

Solving this probability for each data point, $x^i$, we get an equation for $\tau_k^i$.

$\tau^i_k = p(z^i = k | x^i, \theta^t) = \frac{p(x^i | z^i = k)p(z^i = k)}{\Sigma_{k'=1..K} p(z^i = k',x^i)}$


Thus, given our equation for the posterior as seen above, we can simplify this expression to be 


$\frac{\pi_k N(x^i | \mu_k,\Sigma_k)}{\Sigma_{k'=1...K} \pi_{k'} N(x^i|\mu_{k'}, \Sigma_{k'}}$


------------------------------

Then, to continue with the E-step, we can now take the function of $f(\theta)$, which is equal to the expectation of the posterior ($E_{q(z^1,z^2,...z^m)}$) times the log-likelihood function ($log \prod^m_{i=1} (p(z^i | x^i,\theta^t)$). We take the log of the function, as it makes the math much easier and nicer, and turns the product into a sum. 

Finally, our last step is to expand the log of the Guassian density function, to get 

$f(\theta) = \Sigma^m_{i=1} E_{p(z^i|x^i,\theta^t)} [log \pi_{z^i} - \frac{1}{2}(x^i - \mu_{z^i})^T \Sigma^{-1}_{z^i}(x^i - \mu_{z^i}) - \frac{1}{2}log|\Sigma_{z^i}| - \frac{n}{2} log(2\pi)]$


Which we can expand into the following function due to being able to substitue the expecation of the posterior for Tau. 

$f(\theta) = \Sigma^m_{i=1} \Sigma^K_{k=1} \tau^i_k [log \pi_k - \frac{1}{2}(x^i - \mu_k)^T \Sigma^{-1}_k (x^i - \mu_k) - \frac{1}{2} log|\Sigma_k| - \frac{n}{2} log(2\pi)]$


------------------------------------

### The M-Step


Now, with our $f(\theta)$ Formalized in the E-step, we then have to maximize $f{\theta}$. In other words, we want to find $\pi_k$, and we know that the sum of all $\pi_k$ for all k = 1.


Thus, to maximize, we can form the Lagrangian, which is a way to formulate an optimization problem without needing constraints, and having a penalty term, $\lambda$, which affects how the equation is maximized. 

The lagrangian is found to be $L = \Sigma^m_{i=1}\Sigma^k_{k=1} \tau^i_k + ....] + \lambda(1-\Sigma^k_{i=1} \pi_k)$

To be able to maximize this, we can simply take the partial derivative, and set equal to 0, to solve for $\pi_k$: 

$\pi_k = \frac{1}{\lambda} \Sigma^m_{i=1} \tau^i_k$

Then, since this sum from k=1 to K of $\pi_k$ = 1, we know that lambda must = m, and thus that $\pi_k = \frac{1}{m} \Sigma^m_{i=1} \tau^i_k$

-------------------------------------------------
### Summary: 

Thus, in summary, we are taking the maximum likelihood to find the best possible distribution for a KMM model. We take 


$\theta$ to be equal to the three parameters of our model PER GAUSSIAN component, in this case ($\pi_k, \mu_k$, and $\Sigma_k$), which are the proportion of values in a given K-component, the mean, and the covariance matrix of that component. 

Thus, to maximize the likelihood, we are taking 

$\theta^*$ = log $\prod^m_{i=1} p(x^i)$, or in other words, maximizing the probability if x, given that it belongs to a specific gaussian component. Since we do not know which gaussian it might belong to, however, or even how the guassian components should be formed in order to best maximize the probability for all x's, we have to take the expectation with respect to our unknown z, and use Bayes rule to reconfigure our original problem, to make it solvable. In other words, given the data points we have, we cycle through taking the expectation, and then based on that expectation, maximizing $f(\theta)$, and then using our new updated parameters to again take the expectation of the posterior using our new values.

## Question b - Implementing EM algorithm

Now, given all that information, I will actully implement the algorithm using python! Again, the bulk of this code comes from the demo code given to us, but I will explain each step with comments.

```python
np.random.seed(9200)
comps = 4
df_decomp, m, VSU = decompose(hm,comps,style='svd')
```

In the code above, I simply am setting the random seed so that my results are reproducible, and use my previously created PCA function, called decompose, to get the output of a matrix that is 1990x4, or in other words, 4 column vectors, which the principle components, with one row per photo.

```python
k = 2 
pi = np.random.randn(k)
pi = pi/np.sum(pi)
mu = np.random.normal(size=(k,comps))
mu_old = mu.copy()
sigma = []
for ii in range(k):
    # to ensure the covariance psd
    # np.random.seed(seed)
    dummy = np.random.normal(size=(comps,comps))
    sigma.append(dummy@dummy.T + np.eye(4))
    
# initialize the posterior
tau = np.full((m, k), fill_value=0.)
```

Because this process is iterative, and we have no way to maximize the entire system in the first step, we must initialize all of our parameters, which, per step in the EM-process, will get updated over every iteration. 

K is set to 2, as we *only want* k gaussian components, as we are testing EM to be able to separate the classifications of 2 digits. 

pi is simply set as two random values from the standard normal distribution, and then normalized by dividing by the sum of both values per value. Mu is also initialized using the standard normal distribution, with size 2x4, to match the amount of gaussian components and principal components in the data. Finally, the covariance matrix, sigma, is also initialized with random values from the standard normal, And tau is initialized as a 1990x2 matrix filled with 0s, whose values will get updated over the EM-iteratives with the probability of a particular data point beloning to the first or second gaussian component.

In the following code below, the EM algorithm is enacted. Notice the max iterations set to 100 -- if the algorithm iterates more than 100 times, then it just automatically stops. In this case, it would be worth looking into why it doesnt converge, or perhaps converges very slowly. Luckily, I do not run into that problem. 

Then, the tolerance is set to a specific small value, as the algorithm may not 100% converge (in which the mu does not change at all), but once it gets reasonably close, the algorithm will stop. 

For the E-step, the math is simply being performed as explained in part A, in which tau is being updated using our initialized variables, and then in the M-step, we are updating our parameters, like pi, mu, and sigma, based on our calculation of tau. In each iteration, the probability of each point belonging to the first or second component is updated, and based on that probability, we then update the parameters, leading to a converging and iterative process that finally groups the points very accurately. 

Note: While this algorithm is using 4 principal components, I still graph each iteration of Principal Components 0 and 1, just to give a visual of what is going on. But remember that the actual data has 4 components.

Finally, not that, in the print out, I only print every other 5'th iteration, but the final line prints the convergence iteration.

```python
maxIter= 100
tol = 1e-3
log_mean_max_tau = []
for ii in range(maxIter):

    # E-step    
    for kk in range(k):
        tau[:, kk] = pi[kk] * mvn.pdf(df_decomp, mu[kk], sigma[kk])
    # normalize tau
    sum_tau = np.sum(tau, axis=1)
    sum_tau.shape = (m,1)    
    tau = np.divide(tau, np.tile(sum_tau, (1, k)))
    
    
    # M-step
    for kk in range(k):
        # update prior
        pi[kk] = np.sum(tau[:, kk])/m
        
        # update component mean
        mu[kk] = df_decomp.T @ tau[:,kk] / np.sum(tau[:,kk], axis = 0)
        
        # update cov matrix
        dummy = df_decomp - np.tile(mu[kk], (m,1)) # X-mu
        sigma[kk] = dummy.T @ np.diag(tau[:,kk]) @ dummy / np.sum(tau[:,kk], axis = 0)
    log_mean_max_tau.append(np.sum(np.log(sum_tau)))
    if ii%5==0:    
        print('-----iteration---',ii)    
        plt.scatter(df_decomp[:,0], df_decomp[:,1], c= tau[:,1])
        plt.axis('scaled')
        plt.draw()
        plt.pause(0.1)
    if np.linalg.norm(mu-mu_old) < tol:
        print(f'training coverged on iteration {ii}')
        break
    mu_old = mu.copy()
    if ii==maxIter-1:
        print(f'max iteration reached: {ii}')
        break
```

As can be seen, the algorithm converged aftr just 16 iterations! Looking at the graphs, even just when graphing components 0 and 1 of the data, we can see that the algorithm did an amazing job in capturing the two sections of the data - the circle and the moon shaped spread. Thus, one of these must relate to the photos of the 2's, and the other to the photos of the 6's. Time to find out which is which! 

Note too how nicely it solves over the iterations. It slowly takes over that circle with the new classification, cleanly leaving the moon-shape cloud of data on its own.


Finally, we can plot the log-likelihood over each iteration to see how the function converged:

```python
plt.plot(log_mean_max_tau[1:])
plt.title("Log-Likelihood Over Each Iteration")
plt.show()
```

As can be seen, the algorithm makes a LOT of progress early on, each iteration making big changes. As the algorithm starts to get close to conversion, only slight changes are made in the parameters, as the probabilities in tau get updated only very slightly, along with the mean and covariance. We can see this behavior a bit in the top photos, as the changes from photo 1 to 2 show big differences, and the changes from photo 3 to 4 seem minimal.
